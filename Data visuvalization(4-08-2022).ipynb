```python
###########LINE PLOT ###########
import matplotlib.pyplot as plt
import numpy as np
plt.title("lineplot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
xpoints=np.array([0,5])
ypoints=np.array([0,100])
plt.plot(xpoints,ypoints)
plt.show()
```
